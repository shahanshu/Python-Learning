```python
#scatter plot to show the relation bet two variabls using dots
x=np.random.rand(22)
y=np.random.rand(22)
plt.scatter(x,y,color='green',label='Randopm points')
plt.show()
```

```python
#plt.bar() is used to compare categorical datas
cat=['a','b','c','d','e']

val=[12,4,6,7,8]
plt.bar(cat,val,color='blue',label='Category values')
plt.show()
```

```python
# histogram = plt.hist()
data=np.random.rand(10)
plt.hist(data,bins=30,color='green',alpha=0.7)
plt.show()
```

```python
#to give a label inside the graphhing/line use 'balel='' and then use legend()


anshu=np.linspace(1,100,10)
y1=np.sin(anshu)
y2=np.cos(anshu)
plt.plot(anshu,y1)
plt.plot(anshu,y2,label='cos(x)')
plt.legend()
plt.show()
```
