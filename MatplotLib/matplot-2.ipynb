```python
fig,ax=plt.subplots(nrows=2,ncols=2)
x=np.random.rand(10)
y=np.random.rand(10)
ax[1,0].plot(y,x,label='line@2')
ax[1,0].set_xlabel('hey')
ax[0,0].plot(x,y,label='line')
ax[0,1].bar(x,y,label='barCharts')
ax[1,1].scatter(x,y,color='red',label='this is dots')
ax[0,1].legend()
plt.tight_layout()
plt.show()
```
