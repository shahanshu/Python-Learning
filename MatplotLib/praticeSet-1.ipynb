```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,50)
y=x**2
plt.plot(y,label="y^x curve",color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
