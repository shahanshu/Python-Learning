```python
x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, color='darkblue')

# Add text
ax.text(0.5, 0.9, 'Peak Point', transform=ax.transAxes, fontsize=12, color='red')

# Find and annotate the max point
max_index = np.argmax(y)
max_x = x[max_index]
max_y = y[max_index]

ax.plot(max_x, max_y, 'ro')  # Highlight the peak
ax.annotate('Max Value', xy=(max_x, max_y), xytext=(max_x + 0.5, max_y),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.title('Sine Wave with Annotation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.style.use('ggplot')
plt.plot(x, y, label='sin(x)')

# Find highest and lowest points
max_idx = np.argmax(y)  # Index of maximum y-value
min_idx = np.argmin(y)  # Index of minimum y-value

# Annotate highest point
plt.annotate(
    f'Max: ({x[max_idx]:.2f}, {y[max_idx]:.2f})',
    xy=(x[max_idx], y[max_idx]),
    xytext=(x[max_idx] + 0.5, y[max_idx] + 0.1),
    arrowprops=dict(facecolor='red', shrink=0.05),
    color='red'
)

# Annotate lowest point
plt.annotate(
    f'Min: ({x[min_idx]:.2f}, {y[min_idx]:.2f})',
    xy=(x[min_idx], y[min_idx]),
    xytext=(x[min_idx] - 1.5, y[min_idx] - 0.2),
    arrowprops=dict(facecolor='blue', shrink=0.05),
    color='blue'
)

plt.legend()
plt.title('Sine Wave with Annotated Peaks')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```
